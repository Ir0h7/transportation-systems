```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = "kasimov_data/opros.csv"
df = pd.read_csv(file_path)

df = df.drop(columns=["ID", "Время создания"], errors="ignore")
total_respondents = len(df)
print("Общее число опрошенных:", total_respondents)

custom_palette = ['#409BFF', '#dae856', '#f13d53', '#ededed',
                  '#eddac4', '#1678e2', '#25253b']

# 1 Пол
values = df['Укажите пол '].value_counts().values
labels = df['Укажите пол '].value_counts().index
wedges, texts, autotexts = plt.pie(
     values,
     labels=labels,
     autopct='%1.1f%%',
     startangle=90,
     colors=custom_palette,
     textprops={'fontsize': 11}
)

for autotext in autotexts:
    autotext.set_fontweight('bold')

plt.title("Пол респондентов", fontsize=14, fontweight='bold')
plt.ylabel("")
plt.show()

# 2 Возраст
plt.figure(figsize=(8, 5))
value_counts = df["Укажите вашу возрастную группу"].value_counts(dropna=False)
value_counts.plot(kind="bar", edgecolor="black", color=custom_palette[0])
plt.title("Возраст респондентов", fontsize=14, fontweight='bold')
plt.xlabel("")
plt.ylabel("Количество ответов")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

percentages = df["Укажите вашу возрастную группу"].value_counts(normalize=True, dropna=False) * 100
result = pd.DataFrame({'Количество': value_counts, 'Процент': percentages.round(1)})
print(result)


plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Укажите вашу возрастную группу', hue='Укажите пол ', palette=custom_palette)
plt.title('Половозрастной состав респондентов', fontsize=14, fontweight='bold')
plt.xlabel('Возрастная группа')
plt.ylabel('Количество респондентов')
plt.xticks(rotation=20, ha='right')
plt.legend(title='Пол')
plt.tight_layout()
plt.show()

# 3 Вы живете в Касимове?
values = df['Вы живете в Касимове?'].value_counts().values
labels = df['Вы живете в Касимове?'].value_counts().index
wedges, texts, autotexts = plt.pie(
     values,
     labels=labels,
     autopct='%1.1f%%',
     startangle=90,
     colors=custom_palette,
     textprops={'fontsize': 11}
)

for autotext in autotexts:
    autotext.set_fontweight('bold')

plt.title("Вы живете в Касимове?", fontsize=14, fontweight='bold')
plt.ylabel("")
plt.show()

# 4 Район проживания
plt.figure(figsize=(8, 5))
value_counts = df["Ваш район проживания (или ближайшая улица):"].value_counts(dropna=False)
value_counts.plot(kind="bar", edgecolor="black", color=custom_palette[0])
plt.title("Район проживания респондентов", fontsize=14, fontweight='bold')
plt.ylabel("Количество ответов")
plt.xlabel("")
plt.xticks(rotation=20, ha="right", fontsize=8)
plt.tight_layout()
plt.show()

percentages = df["Ваш район проживания (или ближайшая улица):"].value_counts(normalize=True, dropna=False) * 100
result = pd.DataFrame({'Количество': value_counts, 'Процент': percentages.round(1)})
print(result)

# 5 У вас есть свой личный транспорт?
values = df['У вас есть свой личный транспорт?'].value_counts().values
labels = df['У вас есть свой личный транспорт?'].value_counts().index
wedges, texts, autotexts = plt.pie(
     values,
     labels=labels,
     autopct='%1.1f%%',
     startangle=90,
     colors=custom_palette,
     textprops={'fontsize': 11}
)

for autotext in autotexts:
    autotext.set_fontweight('bold')

plt.title("У вас есть свой личный транспорт?", fontsize=14, fontweight='bold')
plt.ylabel("")
plt.show()

# 6 Как часто вы пользуетесь личным транспортом?
plt.figure(figsize=(8, 5))
value_counts = df["Как часто вы пользуетесь личным транспортом?"].value_counts(dropna=False)
value_counts.plot(kind="bar", edgecolor="black", color=custom_palette[0])
plt.title("Как часто вы пользуетесь личным транспортом?", fontsize=14, fontweight='bold')
plt.xlabel("")
plt.ylabel("Количество ответов")
plt.xticks(rotation=20, ha="right", fontsize=8)
plt.tight_layout()
plt.show()

percentages = df["Как часто вы пользуетесь личным транспортом?"].value_counts(normalize=True, dropna=False) * 100
result = pd.DataFrame({'Количество': value_counts, 'Процент': percentages.round(1)})
print(result)

# 7 Вы пользуетесь общественным транспортом?
values = df['Вы пользуетесь общественным транспортом?'].value_counts().values
labels = df['Вы пользуетесь общественным транспортом?'].value_counts().index
wedges, texts, autotexts = plt.pie(
     values,
     labels=labels,
     autopct='%1.1f%%',
     startangle=90,
     colors=custom_palette,
     textprops={'fontsize': 11}
)

for autotext in autotexts:
    autotext.set_fontweight('bold')

plt.title("Вы пользуетесь общественным транспортом?", fontsize=14, fontweight='bold')
plt.ylabel("")
plt.show()

# 8 Как часто вы пользуетесь общественным транспортом?
plt.figure(figsize=(8, 5))
value_counts = df["Как часто вы пользуетесь общественным транспортом?"].value_counts(dropna=False)
value_counts.plot(kind="bar", edgecolor="black", color=custom_palette[0])
plt.title("Как часто вы пользуетесь общественным транспортом?", fontsize=14, fontweight='bold')
plt.ylabel("Количество ответов")
plt.xlabel("")
plt.xticks(rotation=20, ha="right", fontsize=8)
plt.tight_layout()
plt.show()

percentages = df["Как часто вы пользуетесь общественным транспортом?"].value_counts(normalize=True, dropna=False) * 100
result = pd.DataFrame({'Количество': value_counts, 'Процент': percentages.round(1)})
print(result)

# 9 Пользуетесь автовокзалом?
values = df['Пользуетесь автовокзалом?'].value_counts().values
labels = df['Пользуетесь автовокзалом?'].value_counts().index
wedges, texts, autotexts = plt.pie(
     values,
     labels=labels,
     autopct='%1.1f%%',
     startangle=90,
     colors=custom_palette,
     textprops={'fontsize': 11}
)

for autotext in autotexts:
    autotext.set_fontweight('bold')

plt.title("Пользуетесь автовокзалом?", fontsize=14, fontweight='bold')
plt.ylabel("")
plt.show()

# 10 Как часто пользуетесь автовокзалом? 
plt.figure(figsize=(8, 5))
value_counts = df["Как часто пользуетесь автовокзалом? "].value_counts(dropna=False)
value_counts.plot(kind="bar", edgecolor="black", color=custom_palette[0])
plt.title("Как часто пользуетесь автовокзалом?", fontsize=14, fontweight='bold')
plt.ylabel("Количество ответов")
plt.xlabel("")
plt.xticks(rotation=20, ha="right", fontsize=8)
plt.tight_layout()
plt.show()

percentages = df["Как часто пользуетесь автовокзалом? "].value_counts(normalize=True, dropna=False) * 100
result = pd.DataFrame({'Количество': value_counts, 'Процент': percentages.round(1)})
print(result)

# 11 Как обычно добираетесь до/от автовокзала?
arrival_methods_cols = [col for col in df.columns if "Как обычно добираетесь до/от автовокзала?" in col]
plt.figure(figsize=(10, 5))
arrival_methods = df[arrival_methods_cols].apply(lambda x: x.dropna().tolist(), axis=1).explode().value_counts()
arrival_methods.plot(kind="bar", edgecolor="black", color=custom_palette[0])
plt.title("Как обычно добираетесь до/от автовокзала?", fontsize=14, fontweight='bold')
plt.ylabel("Количество ответов")
plt.xticks(rotation=20, ha="right", fontsize=10)
plt.tight_layout()
plt.show()

percentages = df[arrival_methods_cols].apply(lambda x: x.dropna().tolist(), axis=1).explode().value_counts(normalize=True) * 100
result = pd.DataFrame({'Количество': arrival_methods, 'Процент': percentages.round(1)})
print(result)

# 12 Какая ваша основная цель пользования автовокзала?
plt.figure(figsize=(6, 6))

# Подсчет значений и процентов
values = df['Какая ваша основная цель пользования автовокзала?'].value_counts()
labels = values.index
sizes = values.values
percents = sizes / sizes.sum() * 100

# Строим круговую диаграмму без подписей на секторах
wedges, texts = plt.pie(
    sizes,
    startangle=90,
    colors=custom_palette,
    wedgeprops=dict(width=0.6, edgecolor="w")
)

# Формируем легенду (название + процент)
legend_labels = [f"{label} — {p:.1f}%" for label, p in zip(labels, percents)]
plt.legend(
    wedges,
    legend_labels,
    title="Варианты ответов",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.title("Какая ваша основная цель пользования автовокзала?", fontsize=14, fontweight='bold')
plt.show()

# 13 Отношение к переносу автовокзала (круговая диаграмма)
values = df['Как вы отнесетесь к переносу автовокзала?'].value_counts().values
labels = df['Как вы отнесетесь к переносу автовокзала?'].value_counts().index
wedges, texts, autotexts = plt.pie(
     values,
     labels=labels,
     autopct='%1.1f%%',
     startangle=90,
     colors=custom_palette,
     textprops={'fontsize': 11}
)

for autotext in autotexts:
    autotext.set_fontweight('bold')

plt.title("Отношение к переносу автовокзала", fontsize=14, fontweight='bold')
plt.ylabel("")
plt.show()

# 14 Оценки сервисов вокзала (средние значения по каждому сервису)
service_cols = [col for col in df.columns if "Оцените важность сервисов на автовокзале (1 — совсем не важно, 5 — обязательно)" in col]
service_means = df[service_cols].mean().sort_values()
service_means.index = service_means.index.str.replace("Оцените важность сервисов на автовокзале (1 — совсем не важно, 5 — обязательно) / ", "")

plt.figure(figsize=(10, 5))
service_means.plot(kind="bar", edgecolor="black", color=custom_palette[0])
plt.title("Средние оценки важности сервисов автовокзала", fontsize=14, fontweight='bold')
plt.ylabel("Средний балл (1–5)")
plt.xticks(rotation=20, ha="right", fontsize=8)
plt.tight_layout()
plt.show()

# 15 Что вы готовы купить/использовать при ожидании около 30 мин?
plt.figure(figsize=(6, 6))
services_to_buy_use_cols = [col for col in df.columns if "Что вы готовы купить/использовать при ожидании около 30 мин?" in col]
services_to_buy_use_cols = df[services_to_buy_use_cols].apply(lambda x: x.dropna().tolist(), axis=1).explode().value_counts()
labels = services_to_buy_use_cols.index
sizes = services_to_buy_use_cols.values
percents = sizes / sizes.sum() * 100

wedges, texts = plt.pie(
    sizes,
    startangle=90,
    colors=plt.cm.tab20.colors,
    wedgeprops=dict(width=0.6, edgecolor="w")
)

legend_labels = [f"{label} — {p:.1f}%" for label, p in zip(labels, percents)]
plt.legend(
    wedges,
    legend_labels,
    title="Варианты ответов",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.title("Что вы готовы купить/использовать при ожидании около 30 мин?", fontsize=14, fontweight='bold')
plt.show()

# 16 Если задержка рейса больше 30 мин, что сделает ваше ожидание комфортней?
plt.figure(figsize=(6, 6))
delayed_travel_cols = [col for col in df.columns if "Если задержка рейса больше 30 мин, что сделает ваше ожидание комфортней?" in col and "(напишите свой вариант)" not in col]
delayed_travel_cols = df[delayed_travel_cols].apply(lambda x: x.dropna().tolist(), axis=1).explode().value_counts()
labels = delayed_travel_cols.index
sizes = delayed_travel_cols.values
percents = sizes / sizes.sum() * 100

wedges, texts = plt.pie(
    sizes,
    startangle=90,
    colors=custom_palette,
    wedgeprops=dict(width=0.6, edgecolor="w")
)

legend_labels = [f"{label} — {p:.1f}%" for label, p in zip(labels, percents)]
plt.legend(
    wedges,
    legend_labels,
    title="Варианты ответов",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.title("Если задержка рейса больше 30 мин, что сделает ваше ожидание комфортней?", fontsize=14, fontweight='bold')
plt.show()

# 18 Сколько вы готовы заплатить за хранение одной сумки/чемодана до 3 часов?
values = df['Сколько вы готовы заплатить за хранение одной сумки/чемодана до 3 часов?'].value_counts().values
labels = df['Сколько вы готовы заплатить за хранение одной сумки/чемодана до 3 часов?'].value_counts().index
wedges, texts, autotexts = plt.pie(
     values,
     labels=labels,
     autopct='%1.1f%%',
     startangle=90,
     colors=custom_palette,
     textprops={'fontsize': 11}
)

for autotext in autotexts:
    autotext.set_fontweight('bold')

plt.title("Сколько вы готовы заплатить за хранение одной сумки/чемодана до 3 часов?", fontsize=14, fontweight='bold')
plt.ylabel("")
plt.show()

# 19 Готовы ли вы пользоваться мобильным приложением вокзала (e-билет, статус рейса)?
values = df['Готовы ли вы пользоваться мобильным приложением вокзала (e-билет, статус рейса)?'].value_counts().values
labels = df['Готовы ли вы пользоваться мобильным приложением вокзала (e-билет, статус рейса)?'].value_counts().index
wedges, texts, autotexts = plt.pie(
     values,
     labels=labels,
     autopct='%1.1f%%',
     startangle=90,
     colors=custom_palette,
     textprops={'fontsize': 11}
)

for autotext in autotexts:
    autotext.set_fontweight('bold')

plt.title("Готовы ли вы пользоваться мобильным приложением вокзала (e-билет, статус рейса)?", fontsize=14, fontweight='bold')
plt.ylabel("")
plt.show()

# 20 Что бы вы хотели видеть на месте автовокзала?
plt.figure(figsize=(6, 6))
station_replace_variants_cols = [col for col in df.columns if "Что бы вы хотели видеть на месте автовокзала?" in col and "(напишите свой вариант)" not in col]
station_replace_variants_cols = df[station_replace_variants_cols].apply(lambda x: x.dropna().tolist(), axis=1).explode().value_counts()
labels = station_replace_variants_cols.index
sizes = station_replace_variants_cols.values
percents = sizes / sizes.sum() * 100

wedges, texts = plt.pie(
    sizes,
    startangle=90,
    colors=plt.cm.tab20.colors,
    wedgeprops=dict(width=0.6, edgecolor="w")
)

legend_labels = [f"{label} — {p:.1f}%" for label, p in zip(labels, percents)]
plt.legend(
    wedges,
    legend_labels,
    title="Варианты ответов",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.title("Что бы вы хотели видеть на месте автовокзала?", fontsize=14, fontweight='bold')
plt.show()
```
